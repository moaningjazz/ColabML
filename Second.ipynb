```
import pandas as pd
import numpy as np
import matplotlib
import matplotlib.pyplot as plt
from sklearn import datasets
from sklearn.model_selection import train_test_split
iris = datasets.load_iris()
%matplotlib inline

data = pd.read_csv('5_OGE_2019.csv', index_col=0)
```

Деление датасета на обучающую и тестовую части
0.9 и 0.1:

```
data_train,data_test = train_test_split(data, train_size=0.9,test_size=0.1, random_state=101)
```

Проверка гипотезы h0 - средняя оценка по всем ученикам:

```
h0 = data_train['Оцека по 5'].mean()
h0 = round(h0)
```

Запись предсказанных оценок из h0 в новый датасет:

```
data_test['Оценка по h0'] = h0
```

Проверка гипотезы h1: медиана по той школе ученика:

```
h1 = data_train.groupby("Краткое наименование ОО")["Оцека по 5"].median()
```

Запиcь предсказанных по h1 оценок в тестовый датасет:

```
data_test = pd.merge(data_test,h1, on ='Краткое наименование ОО')
data_test = data_test.rename(columns = {'Оцека по 5_x' : 'Оцека по 5', 'Оцека по 5_y' : 'Оценка по h1'})
data_test['Оценка по h1'] = round(data_test['Оценка по h1']) 
```

Проверка гипотезы h2: мода по всем школам этого типа:

```
h2 = (data_train.groupby("Тип ОО")["Оцека по 5"]).agg(lambda x: pd.Series.mode(x)[0])
```

Запиcь предсказанных по h2 оценок в тестовый датасет:

```
data_test = pd.merge(data_test,h2, on ='Тип ОО')
data_test = data_test.rename(columns = {'Оцека по 5_x' : 'Оцека по 5', 'Оцека по 5_y' : 'Оценка по h2'})
data_test['Оценка по h2'] = round(data_test['Оценка по h2']) 
```

Проверка гипотезы h3: медиана по варианту:

```
h3 = data_train.groupby("Вариант")["Оцека по 5"].median()
```

Запись предсказанных по h3 оценкок в тестовый датасет:

```
data_test = pd.merge(data_test,h3, on ='Вариант')
data_test = data_test.rename(columns = {'Оцека по 5_x' : 'Оцека по 5', 'Оцека по 5_y' : 'Оценка по h3'})
```

Проверка точности предсказывания по h0 в процентах:

```
data_test['Погрешность h0'] = (data_test['Оцека по 5'] - data_test['Оценка по h0']).map(lambda x: abs(x))
full_size = data_test['Оцека по 5'].count()
right_size_h0 = data_test[data_test["Погрешность h0"] == 0]["Оцека по 5"].count()
percent_h0 = (right_size_h0 / full_size) * 100
print("Доля верно угаданных по h0 оценок:")
print(percent_h0)
```

Проверка точности предсказывания по h0 (ошибочный балл на человека):

```
prediction_shift_h0 = data_test['Погрешность h0'].sum()
prediction_shift_h0 /= full_size
print("Погрешность верно угаданных оценок по h0:")
print(prediction_shift_h0)
```

Проверка точности предсказывания по h1 (в процентах):

```
data_test['Погрешность h1'] = (data_test['Оцека по 5'] - data_test['Оценка по h1']).map(lambda x: abs(x))
right_size_h1 = data_test[data_test["Погрешность h1"] == 0]["Оцека по 5"].count()
percent_h1 = (right_size_h1 / full_size) * 100
print("Доля верно угаданных по h1 оценок:")
print(percent_h1)
```

Проверка точности предсказывания по h1 (ошибочный балл на человека):

```
prediction_shift_h1 = data_test['Погрешность h1'].sum()
prediction_shift_h1 /= full_size
print("Погрешность верно угаданных оценок по h1:")
print(prediction_shift_h1)
```

Проверка точности предсказывания по h2 (в процентах):

```
data_test['Погрешность h2'] = (data_test['Оцека по 5'] - data_test['Оценка по h2']).map(lambda x: abs(x))
right_size_h2 = data_test[data_test["Погрешность h2"] == 0]["Оцека по 5"].count()
percent_h2 = (right_size_h2 / full_size) * 100
print("Доля верно угаданных по h2 оценок:")
print(percent_h2)
```
