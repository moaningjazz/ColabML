```
import pandas as pd
from graphviz import Source
from IPython.display import display, HTML, SVG
from sklearn import tree
from sklearn.model_selection import train_test_split
style = "<style>svg{width:70% !important; height:70% !important;}</style>"
HTML(style)

clf = tree.DecisionTreeClassifier(max_depth=9)

titanic_data = pd.read_csv('titanic_train.csv')
X = titanic_data.drop(['PassengerId', 'Survived', 'Name', 'Ticket', 'Cabin'], axis=1)
y = titanic_data.Survived
X = pd.get_dummies(X)
X = X.fillna({'Age': X.Age.median()})
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)
clf.fit(X_train, y_train)
graph = Source(tree.export_graphviz(clf, out_file=None, feature_names=list(X), class_names=['Died', 'Survived'], filled=True))
display(SVG(graph.pipe(format='svg')))
print("Значение качества: " + str(clf.score(X_train, y_train) - clf.score(X_test, y_test)))
```

3) Сочетание первого и второго с добавлением ограничения на число листьев max_leaf_nodes=2, качество улучшилось

```
import pandas as pd
from graphviz import Source
from IPython.display import display, HTML, SVG
from sklearn import tree
from sklearn.model_selection import train_test_split
style = "<style>svg{width:70% !important; height:70% !important;}</style>"
HTML(style)

clf = tree.DecisionTreeClassifier(max_features='log2', max_depth=2, max_leaf_nodes=2)

titanic_data = pd.read_csv('titanic_train.csv')
X = titanic_data.drop(['PassengerId', 'Survived', 'Name', 'Ticket', 'Cabin'], axis=1)
y = titanic_data.Survived
X = pd.get_dummies(X)
X = X.fillna({'Age': X.Age.median()})
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)
clf.fit(X_train, y_train)
graph = Source(tree.export_graphviz(clf, out_file=None, feature_names=list(X), class_names=['Died', 'Survived'], filled=True))
display(SVG(graph.pipe(format='svg')))
print("Значение качества: " + str(clf.score(X_train, y_train) - clf.score(X_test, y_test)))
```
