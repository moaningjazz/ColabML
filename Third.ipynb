```
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt 
from scipy import stats
%matplotlib inline

data = pd.read_csv('ABBREV.csv', index_col=0)

data
```

```
data = pd.read_csv('ABBREV.csv', sep=';', encoding='cp1251')
data
```

Общая статистика

```
print('Средние содержание протеина = '+ str(np.mean(data['Protein_(g)'])))
print('Максимальное содержание протеина = '+ str(np.max(data['Protein_(g)'])))
print('Минимальное содержание протеина = '+ str(np.min(data['Protein_(g)'])))
print('Медиана = '+ str(np.median(data['Protein_(g)'])))
```

Деление на две группы по медиане

```
temp2=data[data['Protein_(g)'] >= np.median(data['Protein_(g)'])]
temp3=data[data['Protein_(g)'] < np.median(data['Protein_(g)'])]
labels = ["1","2"]
values = [ len(temp2), len(temp3)]
explode = (0.1, 0.1)
plt.pie(values, labels=labels, autopct='%1.3f%%', shadow = True, explode = explode, startangle = 90)
plt.axis('equal')
plt.show()
```

Деление по калорийности

Низкокалорийные до 100, умереннокалорийные (можно объединить и до 400 обе группы), высококалорийные 

```
temp1=data.loc[(data['Energ_Kcal'] == 100)]
temp2=data.loc[(data['Energ_Kcal'] > 101) & (data['Energ_Kcal'] < 200)]
temp3=data.loc[(data['Energ_Kcal'] >= 201) & (data['Energ_Kcal'] < 500)]
temp4=data.loc[(data['Energ_Kcal'] >= 501)]
labels = ["Низко","Умеренно","Умеренно побольше","Высоко"]
values = [ len(temp1),len(temp2), len(temp3),len(temp4)]
explode = (0.2, 0.2, 0.2, 0.2)
plt.pie(values, labels = labels, autopct = '%1.3f%%', shadow = True, explode = explode)
plt.axis('equal')
plt.show()
```
